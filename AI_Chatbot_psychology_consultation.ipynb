```python
import pandas as pd

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

### model load

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')
```

```python
hi1 = model.encode('너는 정말 나빠')
hi2 = model.encode('나는 정말 나빠')
```

```python
cosine_similarity([hi1], [hi2])
```

```python
hi1 = model.encode('너 는 정 말 나 빠')
hi2 = model.encode('너는 정말 나빠')
```

```python
cosine_similarity([hi1], [hi2])
```

### data load

```python
df = pd.read_csv('https://github.com/kairess/mental-health-chatbot/raw/master/wellness_dataset_original.csv')
df.info()
```

```python
df.head(10)
```

```python
df = df.drop(columns='Unnamed: 3')
```

```python
df = df.dropna()
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df.loc[0, '유저']
```

```python
model.encode(df.loc[0, '유저'])
```

```python
df['embedding'] = pd.Series([[]] * len(df)) # dummy

df['embedding'] = df['유저'].map(lambda x: list(model.encode(x)))

df.head()
```

```python
text = '요즘 남편이 비트코인도 하고 너무 속을 썩이네'

embedding_text = model.encode(text)

embedding_text
```

```python
df['similarity'] = df['embedding'].map(lambda x: cosine_similarity([embedding_text], [x]).squeeze())

df.head()
```

```python
df.loc[df['similarity'].idxmax()]['챗봇']
```

```python
answer = df.loc[df['similarity'].idxmax()]

print(answer['구분'])
print(answer['유저'])
print(answer['챗봇'])
print(answer['similarity'])
```
